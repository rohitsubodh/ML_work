```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

#  Xis the 10x10 Hilbert matrix
X=1./(np.arange(1,11) + np.arange(0,10)[:,np.newaxis])
y=np.ones(10)

# ####################################################################
# compute path

n_alphas=200
alphas=np.logspace(-10,-2,n_alphas)

coefs=[]
for a in alphas:
    ridge=linear_model.Ridge(alpha=a,fit_intercept=False)
    ridge.fit(X,y)
    coefs.append(ridge.coef_)
# ##################################################################
# Display results
plt.figure(figsize=(10,8))
ax=plt.gca()

ax.plot(alphas,coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])#reverse axis
plt.xlabel('alpha')
plt.ylabel('weight')
plt.title('Ridge coefficient as a function of the regularization')
plt.axis('tight')
plt.show()


#as alpha increase regulalisation
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')

%matplotlib inline
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)
n_sample=100
rng= np.random.randn(n_sample)*10  #randn generate no. b/w(0,1)
y_gen=0.5*rng+2*np.random.randn(n_sample)

lr=LinearRegression()
lr.fit(rng.reshape(-1,1),y_gen)
model_pred = lr.predict(rng.reshape(-1,1))

plt.figure(figsize=(10,8));
plt.scatter(rng,y_gen);
plt.plot(rng,model_pred);
print("coefficient estimate",lr.coef_)
```

```python
rng.shape
```

```python
idx=rng.argmax() #MAXIMUM value of x or(rng)
#y_gen[idx]
y_gen[idx]=200
idx2=rng.argmin() #MAXIMUM value of x or(rng)
#y_gen[idx]
y_gen[idx2]=-200
```

```python
plt.figure(figsize=(12,10))
plt.scatter(rng,y_gen)

o_lr_model=LinearRegression(normalize=True)
o_lr_model.fit(rng.reshape(-1,1),y_gen)
o_model_pred=o_lr_model.predict(rng.reshape(-1,1))

#plt.scatter(rng,y_gen)
plt.plot(rng,o_model_pred);
print("coefficient estimate",o_lr_model.coef_)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
plt.figure(figsize=(10,8));

ridge_mod = Ridge(alpha=1,normalize=True)
ridge_mod.fit(rng.reshape(-1,1),y_gen)
ridge_pred=ridge_mod.predict(rng.reshape(-1,1))

plt.scatter(rng,y_gen)
plt.plot(rng,ridge_pred);
print("coefficient estimate",ridge_mod.coef_);
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
lasso_mod=Lasso(alpha=.4,normalize=True)
lasso_mod.fit(rng.reshape(-1,1),y_gen)
lasso_pred=lasso_mod.predict(rng.reshape(-1,1))
plt.figure(figsize=(12,10))
plt.scatter(rng.reshape(-1,1),y_gen.reshape(-1,1))
plt.plot(rng,lasso_pred);
print("coefficent estimater",lasso_mod.coef_)
```

# Elastic Net Regression

```python
from sklearn.linear_model import ElasticNet
```

```python
en_mod=ElasticNet(alpha=.02,normalize=True)
en_mod.fit(rng.reshape(-1,1),y_gen)
en_pred=en_mod.predict(rng.reshape(-1,1))


plt.figure(figsize=(12,10))
plt.scatter(rng.reshape(-1,1),y_gen.reshape(-1,1))
plt.plot(rng,en_pred);
print("coefficent estimater",en_mod.coef_)
```
